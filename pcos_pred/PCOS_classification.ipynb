```python
# Replace any non-numeric values in the 'AMH(ng/mL)' column with NaN
df_new['AMH(ng/mL)'] = pd.to_numeric(df_new['AMH(ng/mL)'], errors='coerce')
df_new['II    beta-HCG(mIU/mL)'] = pd.to_numeric(df_new['II    beta-HCG(mIU/mL)'], errors='coerce')

# Now convert the column to float
df_new['AMH(ng/mL)'] = df_new['AMH(ng/mL)'].astype(float)
df_new['II    beta-HCG(mIU/mL)'] = df_new['II    beta-HCG(mIU/mL)'].astype(float)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
imputer.fit(df_new)
df_new = pd.DataFrame(imputer.transform(df_new), columns=df_new.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaling=StandardScaler()
```

```python
scaled_df=scaling.fit_transform(df_new)
```

```python
from sklearn.decomposition import PCA

principal=PCA(n_components=3)
principal.fit(scaled_df)
x=principal.transform(scaled_df)
```

```python
x
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

labels = df_new['PCOS (Y/N)']

model = LogisticRegression(max_iter=200)
rfe = RFE(estimator=model, n_features_to_select=10)
rfe.fit(scaled_df, labels)
```

```python
selected_features = df_new.columns[rfe.support_]
feature_ranking = pd.Series(rfe.ranking_, index=df_new.columns)

selected_features, feature_ranking
```

```python
X = pd.DataFrame(scaled_df, columns=df_new.columns)[['Cycle(R/I)', 'AMH(ng/mL)', 'Weight gain(Y/N)',
            'hair growth(Y/N)', 'Skin darkening (Y/N)', 'Pimples(Y/N)',
            'Fast food (Y/N)', 'Follicle No. (L)', 'Follicle No. (R)']]

y = df_new['PCOS (Y/N)']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
lreg=LogisticRegression()
lreg.fit(X_train,y_train)
```

```python
y_pred=lreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
